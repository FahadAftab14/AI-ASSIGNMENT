```python
import random

def first_choice_hc(landscape, start):
    current = start
    path = [current]

    while True:
        moved = False
        current_value = landscape[current]

        if current > 0:
            if landscape[current - 1] > current_value:
                current -= 1
                path.append(current)
                moved = True

        if not moved and current < len(landscape) - 1:
            if landscape[current + 1] > current_value:
                current += 1
                path.append(current)
                moved = True

        if not moved:
            break

    return path, current


def stochastic_hc(landscape, start):
    current = start
    path = [current]

    while True:
        current_value = landscape[current]
        neighbors = []

        if current > 0 and landscape[current - 1] > current_value:
            neighbors.append(current - 1)

        if current < len(landscape) - 1 and landscape[current + 1] > current_value:
            neighbors.append(current + 1)

        if not neighbors:
            break

        current = random.choice(neighbors)
        path.append(current)

    return path, current

def find_local_maxima(landscape):
    maxima = []
    for i in range(len(landscape)):
        left = landscape[i - 1] if i > 0 else float('-inf')
        right = landscape[i + 1] if i < len(landscape) - 1 else float('-inf')

        if landscape[i] > left and landscape[i] > right:
            maxima.append(i + 1)
    return maxima


def random_restart_hc(landscape, num_restarts, variant='first choice'):
    results = []
    best_state = None
    best_value = float('-inf')

    for r in range(num_restarts):
        start = random.randint(0, len(landscape) - 1)

        if variant == 'first choice':
            path, terminal = first_choice_hc(landscape, start)
        else:
            path, terminal = stochastic_hc(landscape, start)

        value = landscape[terminal]

        results.append((start, terminal, path))

        if value > best_value:
            best_value = value
            best_state = terminal

    return best_state, best_value, results


def run_rrhc_demo():
    landscape = [5, 8, 6, 12, 9, 7, 17, 14, 10, 6, 19, 15, 11, 8]

    print("Local Maxima:", find_local_maxima(landscape))

    for variant in ['first choice', 'stochastic']:
        print(f"\n===== {variant.upper()} =====")

        best_state, best_value, results = random_restart_hc(landscape, 20, variant)

        print(f"{'Restart':<8}{'Start':<8}{'Terminal':<10}{'Value':<8}{'Global?'}")

        for i, (start, terminal, path) in enumerate(results):
            is_global = "Yes" if terminal == 10 else "No"
            print(f"{i+1:<8}{start+1:<8}{terminal+1:<10}{landscape[terminal]:<8}{is_global}")

        print("Best State:", best_state + 1, "Value:", best_value)


def experiment_probabilities():
    landscape = [5, 8, 6, 12, 9, 7, 17, 14, 10, 6, 19, 15, 11, 8]
    restart_values = [1, 3, 5, 10, 20]

    print("\nEmpirical Probabilities:\n")

    for n in restart_values:
        success = 0

        for _ in range(100):
            best_state, _, _ = random_restart_hc(landscape, n)

            if best_state == 10:
                success += 1

        prob = success / 100
        print(f"Restarts={n}, Probability={prob:.2f}")


def plateau_experiment():
    landscape = [5, 8, 6, 12, 9, 7, 17, 17, 10, 6, 19, 15, 11, 8]

    plateau_count = 0
    success = 0

    for _ in range(20):
        start = random.randint(0, len(landscape) - 1)
        _, terminal = first_choice_hc(landscape, start)

        if terminal in [6, 7]:  # states 7,8
            plateau_count += 1

        if terminal == 10:
            success += 1

    print("\nPlateau Experiment:")
    print("Reached Global Max:", success)
    print("Stuck on Plateau:", plateau_count)


def main():
    print("\n===== PART (a) =====")
    run_rrhc_demo()

    print("\n===== PART (b) =====")
    experiment_probabilities()

    print("\n===== PART (c) =====")
    plateau_experiment()

main()
```
